# Proyecto k means

## Ejemplo 1

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Datos sintticos
data = np.random.rand(100, 2)  # Generar datos aleatorios

# Crear el modelo de KMeans con k=3
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)

# Obtener las etiquetas de los clusters y los centroides
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Graficar los clusters y los centroides
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', c='red', s=200)
plt.title('Agrupamiento K-Medias')
plt.xlabel('Caracterstica 1')
plt.ylabel('Caracterstica 2')
plt.show()
```

## Ejemplo 2

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generacin de datos sintticos para partculas con energa y momentum
# Datos simulados con distribuciones normales
np.random.seed(0)
particles = np.concatenate([np.random.normal(5, 1, (100, 2)),  # Grupo 1
                            np.random.normal(1, 0.5, (100, 2)),  # Grupo 2
                            np.random.normal(9, 1.5, (100, 2))])  # Grupo 3

# Aplicacin del algoritmo KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(particles)

# Obtencin de los resultados
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Graficar los resultados
plt.scatter(particles[:, 0], particles[:, 1], c=labels, cmap='viridis', alpha=0.7)
plt.scatter(centers[:, 0], centers[:, 1], marker='*', c='red', s=200)
plt.xlabel('Energa')
plt.ylabel('Momentum')
plt.title('Clasificacin de Partculas en un Acelerador de Partculas')
plt.show()
```

## Ejemplo 3

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Cargar los datos de terremotos
data = pd.read_csv('earthquake_data.csv')

# Seleccionar las columnas relevantes para el clustering (p. ej., magnitud, profundidad, etc.)
selected_columns = ['Magnitude', 'Depth', 'Latitude', 'Longitude']
earthquake_data = data[selected_columns]

# Escalar los datos para estandarizar las caractersticas
scaler = StandardScaler()
scaled_data = scaler.fit_transform(earthquake_data)

# Aplicar K-Means con un nmero arbitrario de clusters
num_clusters = 5  # Nmero de clusters deseado
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(scaled_data)

# Obtener las etiquetas de los clusters y asignarlas a los datos
cluster_labels = kmeans.labels_

# Aadir las etiquetas de clusters al conjunto de datos original
earthquake_data['Cluster'] = cluster_labels

# Visualizacin de los clusters
plt.scatter(earthquake_data['Longitude'], earthquake_data['Latitude'], c=earthquake_data['Cluster'], cmap='viridis')
plt.title('Clusters de Terremotos')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.show()
```
